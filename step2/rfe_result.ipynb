```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('font',family='Times New Roman')
# 设置
pd.options.display.notebook_repr_html=False  # 表格显示
plt.rcParams['figure.dpi'] = 300  # 图形分辨率
sns.set_theme(style='white')  # 图形主题
sns.palplot(sns.color_palette("OrRd_r"))
```

```python
# 含有置信区间的折线图
plt.rc('font',family='Times New Roman')
sns.scatterplot(x="feature_count",
                y="value_auc",
                data=dd,c=plt.get_cmap('Paired')(2))

sns.lineplot(data=dd,x='feature_count',y='value_auc',ci=95,color=plt.get_cmap('Paired')(0))
plt.axvline(32,linestyle='--',linewidth = '1',color=plt.get_cmap('Paired')(5))

plt.xlabel('dropped_feature_count', fontsize=15)
plt.ylabel('value_auc', fontsize=15)
plt.savefig('rfe.pdf',dpi=900,bbox_inches='tight')
plt.show()
```
